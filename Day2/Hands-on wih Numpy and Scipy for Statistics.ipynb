```python
import numpy as np
```

```python
# Normal Distribution
np.random.seed(42)
normal_data=np.random.normal(loc=0, scale=1, size=1000)
print("Mean:", np.mean(normal_data))
print("Median:", np.median(normal_data))
print("Standard deviation:",np.std(normal_data))
print("Variance:",np.var(normal_data))
```

```python
# Uniform distribution
uniform_data=np.random.randint(low=1, high=11, size=500) # high is exclusive
print(uniform_data)
print("Mean:", np.mean(uniform_data))
print("Median:", np.median(uniform_data))
print("Standard deviation:",np.std(uniform_data))
print("Variance:",np.var(uniform_data))
```

```python
import matplotlib.pyplot as plt

plt.hist(normal_data, bins=20, density=True, alpha=0.7, color='skyblue')
plt.title("Normal Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.hist(uniform_data, bins=10, density=True, alpha=0.7, color='skyblue')
plt.title("Uniform Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
from scipy import stats
```

```python
# Normal Distribution probability(less than 1)
prob_normal_lt_1=stats.norm.pdf(1,loc=0, scale=1)
print("Probability (Normal<1):", prob_normal_lt_1)
```

```python
# Uniform distribution probability (between 3 and 7 inclusive)
prob_uniform_3_to_7 = stats.uniform.cdf(7, loc=1, scale=9) - stats.uniform.cdf(2, loc=1, scale=9) # scale is high - low
print("Probability (Uniform between 3 and 7):", prob_uniform_3_to_7)
```
